Data Cleaning

```python
reviews = []
for i in wine['quality']:
    if i >= 1 and i <= 3:
        reviews.append('1')
    elif i >= 4 and i <= 7:
        reviews.append('2')
    elif i >= 8 and i <= 10:
        reviews.append('3')
wine['reviews'] = reviews
```

```python
wine_copy = wine.copy()
```

```python
wine_copy.head(3)
```

```python
wine_copy.drop('color' , axis = 1, inplace=True)
wine_copy.drop('quality' , axis = 1, inplace=True)
wine_copy.head(3)
```

```python
X = wine_copy.drop('reviews', axis = 1).values
y = wine_copy['reviews'].values
print(X.shape)
```

Preprocessing

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
print(X.shape)
```

PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X)
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
```

```python
pca_new = PCA(n_components=8)
X = pca_new.fit_transform(X)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
import warnings
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='gini', random_state=42)
```

```python
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
adb = AdaBoostClassifier(base_estimator=rfc, n_estimators=100)
adb.fit(X_train, y_train)
adb_pred = adb.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
gdb = GradientBoostingClassifier(n_estimators=300, subsample=0.8, max_depth=1)
gdb.fit(X_train, y_train)
gdb_pred = gdb.predict(X_test)
print(classification_report(y_test, y_pred))
```

Support Vector machines

```python
from sklearn.svm import SVC
svc  = SVC()
svc.fit(X_train, y_train)
y_pred  = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
```

```python
gcv = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=10)
gcv.fit(X_train, y_train)
gcv_pred = gcv.predict(X_test)
print(gcv.best_params_)
print(gcv.best_estimator_)
```

```python
svc_params = SVC(C = 1.4, gamma = 1.4, kernel = 'rbf')
svc_params.fit(X_train, y_train)
svc_params_pred = svc_params.predict(X_test)
print(classification_report(y_test, svc_params_pred))
```
